# Data Exploration

## Imports

```python
import zstandard as zstd
import pandas as pd
```

## Sourcing

```python
file_path = "../data/ais_2025_01_01.csv.zst"
```

```python
df = None

with open(file_path, "rb") as f:
    dctx = zstd.ZstdDecompressor()
    with dctx.stream_reader(f) as reader:
        df = next(pd.read_csv(reader, chunksize=100000))
```
